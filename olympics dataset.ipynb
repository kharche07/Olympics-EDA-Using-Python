```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# LOAD DATASET
athletes = pd.read_csv('E:/Analysis Master course/E.D.A/athlete_events.csv')
region = pd.read_csv('E:/Analysis Master course/E.D.A/noc_regions.csv')
```

```python
# JOIN THE DATAFRAMES
athletes_df = athletes.merge(region, how='left', on ='NOC')
athletes_df.head()
```

```python
#COLUMN NAMES CONSISTENT
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True);
```

```python
# Top countries contributing
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
# Plot for 10 countries
plt.figure(figsize=(12,6))
plt.title('Overall Contribution by Country')
sns.barplot(x=top_10_countries.index, y= top_10_countries, palette ='Set2');
```

```python
# Age distribution of Athletes
plt.figure(figsize=(12,6))
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number Of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color ='blue', edgecolor ='white');
```

```python
# winter sports
winter_sports= athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# summer olympics sport
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# pie plot for male and female participants
plt.figure(figsize=(12,6))
plt.title ='Gender_Distribution'
plt.pie(gender_counts, labels = gender_counts.index, autopct ='%1.1f%%', startangle=150, shadow=True);
```

```python
# total medals
total_medals = athletes_df.Medal.value_counts()
total_medals
```

```python
#total number of females in each olympics
female_participants = athletes_df[(athletes_df.Sex =='F') & (athletes_df.Season == 'Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
# total number of male participants in each olympics
male_participants = athletes_df[(athletes_df.Sex =='M')][['Sex','Year']]
male_participants = male_participants.groupby('Year').count().reset_index()
male_participants.head(7)
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season =='Summer')] 
womenOlympics.head()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data = womenOlympics, palette ="Spectral")
plt.title = ('Women Participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'F'].plot()
plt.title = ("PLot of Female Athletes over time")
```

```python
# Gold Medal Atheletes
goldMedals = athletes_df[athletes_df.Medal == 'Gold']
goldMedals.head()
```

```python
# take only values which are different from Nan
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
goldMedals.head()
```

```python
# gold beyond 60
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
# plot for sporting event
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title=("Gold Medals for Age over 60 years")
```

```python
# gold medal in each country

goldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals =goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g =sns.catplot(x='index',y='Medal',data= totalGoldMedals,height=5,kind = 'bar',palette='rocket')
g.despine(left=True)
g.set_xlables=("Top 5 countries")
g.set_ylables=("Number of Medals")
plt.title=("Gold Medals with each country")
```
